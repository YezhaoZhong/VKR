# DGI

```python
datas1 = [
        KR1_roc, 
        KRNMF1_roc, 
        LNSM_WMK1_jaccard_roc,
        LNSM_WMK1_RLN_roc,
        naive_roc,
        SVML1_roc,
        SVMRBF1_roc,
        SVMWGTS1_roc
        ]

datas2 = [
        KR1_pr, 
        KRNMF1_pr, 
        LNSM_WMK1_jaccard_pr,
        LNSM_WMK1_RLN_pr,
        naive_pr,
        SVML1_pr,
        SVMRBF1_pr,
        SVMWGTS1_pr
        ]

names = [
         "KR",
         "VKR",
         "LNSM_jaccard",
         "LNSM_RLN",
         "Naive",
         "SVM_linear",
         "SVM_RBF",
         "SVM_WGTS"
        ]

colors = [
        #  "#9E9E9E",
         "darkred",
         "red",
         "#B883D4",
         "#CFEAF1",
         "black",
         "#C4A5DE",
         "#A1A9D0",
         "#F6CAE5",
         "#96CCCB",
        #  "#C76DA2",
        #  "#8983BF",
        #  "#9DC3E7",
        ]

plt.style.use("seaborn-whitegrid")

AUPR = []
AUROC = []
for data1, data2 in zip(datas1, datas2):
    fpr = data1.iloc[0,:]
    tpr = data1.iloc[1,:]
    
    pr = data2.iloc[0,:]
    rc = data2.iloc[1,:]
    AUPR.append(auc(rc, pr))
    AUROC.append(auc(fpr, tpr))

sort_idx = np.argsort(-np.array(AUROC))
names1_ranked = [names[i] for i in sort_idx]
datas1_ranked = [datas1[i] for i in sort_idx]
colors1_ranked = [colors[i] for i in sort_idx]

plt.figure(figsize=(5,5), dpi=350)
for name, data, colorname in zip(names1_ranked, datas1_ranked, colors1_ranked):
    # print(data)
    fpr = data.iloc[0,:]
    tpr = data.iloc[1,:]
    plt.plot(fpr, tpr, lw=1.2, label='{} (AUC={:.2f})'.format(name, auc(fpr, tpr)),color = colorname, linestyle='--')
    plt.axis('square')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC - methods with feature DGI')
    plt.legend(loc='lower right',fontsize=10)
plt.plot([0, 1], [0, 1], '--', lw=1, color = 'grey')
plt.savefig(fname="overall_ROC_DGI.jpg",bbox_inches="tight")
plt.show()

sort_idx = np.argsort(-np.array(AUPR))
names2_ranked = [names[i] for i in sort_idx]
datas2_ranked = [datas2[i] for i in sort_idx]
colors2_ranked = [colors[i] for i in sort_idx]


plt.figure(figsize=(5,5), dpi=350)
for name, data, colorname in zip(names2_ranked, datas2_ranked, colors2_ranked):
    # print(data)
    pr = data.iloc[0,:]
    rc = data.iloc[1,:]
    plt.plot(rc, pr, lw=1.2, label='{} (AUC={:.2f})'.format(name, auc(rc, pr)),color = colorname, linestyle='--')
    plt.axis('square')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('Recall')
    plt.ylabel('Precsion')
    plt.title('PR - methods with feature DGI')
    plt.legend(loc='upper right',fontsize=10)
plt.plot([0, 1], [1, 0], '--', lw=1, color = 'grey')
plt.savefig(fname="overall_PR_DGI.jpg",bbox_inches="tight")
plt.show()


    # print(newAUC)
# AUPR_normalized = (AUPR - np.mean(AUPR)) / np.std(AUPR)
# AUROC_normalized = (AUROC - np.mean(AUROC)) / np.std(AUROC)
# newAUC = AUROC_normalized + AUPR_normalized

newAUC = np.array(AUROC) + np.array(AUPR)
sort_idx = np.argsort(-newAUC)

plt.figure(figsize=(5,5), dpi=350)
plt.bar(np.array(names)[sort_idx], newAUC[sort_idx], color = np.array(colors)[sort_idx])
# plt.axis('square')
xaxis = np.arange(0, 0+len(newAUC), 1)
for x,y in zip(xaxis, newAUC[sort_idx].tolist()):
    plt.text(x-0.18, y-0.09 , '{:.3f}'.format(y), rotation=90, color='white')
plt.hlines(newAUC[4], xmin=-0.5, xmax=7.5, linestyles="--", lw=1.5, colors="black")

plt.ylim([0.8, 1.4])
plt.xticks(rotation = 90)
plt.title('AUPR + AUROC with feature DGI')
plt.savefig(fname="newAUC_DGI.jpg",bbox_inches="tight")
plt.show()
```
