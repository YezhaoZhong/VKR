## VKR

```python
datas1 = [
        KRNMF1_roc, 
        KRNMF2_roc, 
        KRNMF12_roc
        ]

datas2 = [
        KRNMF1_pr, 
        KRNMF2_pr, 
        KRNMF12_pr
        ]

names = [
         "VKR_DGI",
         "VKR_Chem",
         "VKR_DGI_Chem"
        #  "LNSM_jaccard",
        #  "LNSM_RLN",
        #  "Naive",
        #  "SVM_linear",
        #  "SVM_RBF",
        #  "SVM_WGTS"
        ]

colors = [
        #  "#9E9E9E",
        #  "darkred",
        #  "red",
         "#B883D4",
         "#CFEAF1",
         "#F0988C",
        #  "#C4A5DE",
        #  "#A1A9D0",
        #  "#F6CAE5",
        #  "#96CCCB",
        #  "#C76DA2",
        #  "#8983BF",
        #  "#9DC3E7",
        ]

plt.style.use("seaborn-whitegrid")
         
         
plt.figure(figsize=(5,5), dpi=350)
for name, data, colorname in zip(names, datas1, colors):
    # print(data)
    fpr = data.iloc[0,:]
    tpr = data.iloc[1,:]
    plt.plot(fpr, tpr, lw=1.5, label='{} (AUC={:.2f})'.format(name, auc(fpr, tpr)),color = colorname, linestyle='--')
    plt.axis('square')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC - VKR')
    plt.legend(loc='lower right',fontsize=8)
plt.plot([0, 1], [0, 1], '--', lw=1, color = 'grey')
plt.savefig(fname="VKR_ROC_DGI_Chem.jpg",bbox_inches="tight")
plt.show()


plt.figure(figsize=(5,5), dpi=350)
for name, data, colorname in zip(names, datas2, colors):
    pr = data.iloc[0,:]
    rc = data.iloc[1,:]
    plt.plot(rc, pr, lw=1.5, label='{} (AUC={:.2f})'.format(name, auc(rc, pr)),color = colorname, linestyle='--')
    plt.axis('square')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('Recall')
    plt.ylabel('Precsion')
    plt.title('PR - VKR')
    plt.legend(loc='upper right',fontsize=8)
plt.plot([0, 1], [1, 0], '--', lw=1, color = 'grey')
plt.savefig(fname="VKR_PR_DGI_Chem.jpg",bbox_inches="tight")
plt.show()
```
